###Regression on the tabular data

You have a dataset (internship_train.csv) that contains 53 anonymized features and a target column. Your task is to build model that predicts a target based on the proposed features. Please provide predictions for internship_hidden_test.csv file. Target metric is RMSE. The main goal is to provide github repository that contains:
jupyter notebook with analysis; 
code for modeling (Python 3); 
file with model predictions; 
readme file;
requirements.txt file.



```python
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, Normalizer
from sklearn.metrics import mean_squared_error
from math import sqrt




```

```python
data = pd.read_csv ('internship_train.csv')   
df = pd.DataFrame(data)
df
```

```python
# поділимо дані на ознаки та їх значення
data, label = np.array(df)[:, 0:53], np.array(df)[:, 53]
data_train, data_test, label_train, label_test = train_test_split(data, label, random_state=0)
```

```python
#застосуємо поліноміальну регресію
lr = linear_model.LinearRegression()
pf2=PolynomialFeatures()   

lr.fit(pf2.fit_transform(data_train) , label_train)
lr_pred = lr.predict(pf2.fit_transform(data_test))
```

```python
#RMSE
mean_squared_error(label_test, lr_pred, squared = False)
```

```python
data = pd.read_csv ('internship_hidden_test.csv')   
dt = pd.DataFrame(data)
dt
```

```python
dt = np.array(dt)
dt
```

```python
fin_pred = lr.predict(pf2.fit_transform(dt))
fin_pred
```

```python
fin_pred = pd.DataFrame(data = fin_pred, columns = ['target'])
fin_pred.to_csv('internship_test_results.csv')
```
